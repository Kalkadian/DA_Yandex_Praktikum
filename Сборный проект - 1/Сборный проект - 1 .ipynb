Введем расшифровку рейтингов:

- E - ("Everyone") — "Для всех": Содержание вполне подходит для всех возрастов. Такие игры могут понравиться и взрослым. Игры с этим рейтингом могут содержать минимальное насилие, в основном «мультяшного» характера. Первоначально "K-A" ("Kids to Adults")

## Гипотезы


Выдвенем и проверим несколько гипотез:
 - Средние пользовательские рейтинги платформ Xbox One и PC одинаковые
 - Средние пользовательские рейтинги жанров Action и Sports разные.


Для проверки обеих гипотез будем использовать t-тест Стьюдента, т.к. в обоих случаях выборки независимы между собой.
Определим пороговое значение alpha = 0.05

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;

- Н0 (нулевая гипотеза) - средние пользовательские рейтинги платформ Xbox One и PC одинаковые.
- Н1 (альтернативная гипотеза) - средние пользовательские рейтинги платформ Xbox One и PC отличаются между собой.

```python
PC = data_rev.query('platform == "PC" and user_score > 0')
Xbox = data_rev.query('platform == "XOne" and user_score > 0')
```

```python
alpha = 0.05

results = st.ttest_ind(
    PC['user_score'], 
    Xbox['user_score'],
    equal_var=False,
    )

print('p-значение:',results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Отклоняем нулевую гипотезу.
Верно следующее утверждение: средние пользовательские рейтинги платформ Xbox One и PC отличаются между собой.

Это можно объяснить качестов исполнения одно и того же тайтла например.Одна и таже игра может выйти на одной из платформ сильно хуже/лучше в техническом плане. Поэтому здесь прослеживается разность оценок. 

UPD: если рассматриваем актуальный период, то здесь оценки уже одинаковые, тоесть качество игр могло улучшится уже к 2016 году,который мы добилию.Разработчики лучше научились оптимизировать игры для обеих платформ. 

### Средние пользовательские рейтинги жанров Action и Sports разные.

 - Н0 (нулевая гипотеза) - средние пользовательские рейтинги жанров Action и Sports одинаковые.
 - Н1 (альтернативная гипотеза) - средние пользовательские рейтинги жанров Action и Sports отличаются между собой.

```python
Sports = data_rev.query('genre == "Sports" and user_score > 0')
Action = data_rev.query('genre == "Action"and user_score > 0')
```

```python
alpha = 0.05

results = st.ttest_ind(
    Sports['user_score'], 
    Action['user_score'],
    equal_var=False,
    )

print('p-значение:',results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Таким образом, остается верным утверждение, что средние пользовательские рейтинги жанров Action и Sports одинаковые.

С чем связана схожесть оценок сложно сказать. Возможно жанры схожи по эмоционалному отклику у игроков.в обоих жанрах действия происходят достоточно схожие, игрок постоянно находится в "движении", возможно здесь связь.

UPD: По актуальным данным мы всеже отвергаем нулевую гипотезу и можем утверждать6 что средние пользовательские рейтинги жанров Action и Sports отличаются между собой.Думаю это связано, что в новом поколении консолей уже начали сильно отличаться друг от друга и давать новый игровой опыт потребителям.
