Описание проекта:

#### распределим по возрастным групам

```python
data_encoded['age_group'] = pd.qcut(data_encoded['age'], q=3, labels=['Low', 'Medium', 'High'])

data_encoded.groupby('age_group')['user_id'].count()
```

```python
# теперь посмотрим минимум и максимум в каждой группе, чтобы корректно определить диапазон групп в названии

def get_age_range(data, age_group):
    age = data.query(f'age_group == "{age_group}"')
    age_min = age['age'].min()
    age_max = age['age'].max()
    return age_min, age_max

age_group = "Low"
min_age, max_age = get_age_range(data_encoded, age_group)
print("Минимальный возраст:", min_age)
print("Максимальный возраст:", max_age)

age_group = "Medium"
min_age, max_age = get_age_range(data_encoded, age_group)
print("Минимальный возраст:", min_age)
print("Максимальный возраст:", max_age)

age_group = "High"
min_age, max_age = get_age_range(data_encoded, age_group)
print("Минимальный возраст:", min_age)
print("Максимальный возраст:", max_age)
```

#### Разделим на группы дохода, чтобы в дальйнешем определить какие группы более отточные

```python
data_encoded['salary_group'] = pd.qcut(data_encoded['est_salary'], q=3, labels=['Low', 'Medium', 'High'])

data_encoded.groupby('salary_group')['user_id'].count()
```

```python
# теперь посмотрим минимум и максимум в каждой группе, чтобы корректно определить диапазон групп в названии

def get_salary_range(data, salary_group):
    salary = data.query(f'salary_group == "{salary_group}"')
    salary_min = salary['est_salary'].min()
    salary_max = salary['est_salary'].max()
    return salary_min, salary_max

salary_group = "Low"
salary_min, salary_max = get_salary_range(data_encoded, salary_group)
print("Минимальный доход:", salary_min)
print("Максимальный доход:", salary_max)

salary_group = "Medium"
salary_min, salary_max  = get_salary_range(data_encoded, salary_group)
print("Минимальный доход:", salary_min)
print("Максимальный доход:", salary_max)

salary_group = "High"
salary_min, salary_max  = get_salary_range(data_encoded, salary_group)
print("Минимальный доход:", salary_min)
print("Максимальный доход:", salary_max)

```

## Статистический тест

### Провери различие дохода

#### Проверим как распределены данные. Применим тест Шапиро-Уилка

```python
#Создадим функцию

def check_normality(sample):
    _,p_value = st.shapiro(sample)
    alpha = 0.05 
    if p_value > alpha:
        print("Распределение данных является нормальным")
    else:
        print("Распределение данных не является нормальным")

```

```python
stay = data_stay['est_salary']
leave = data_leave['est_salary']
```

```python
check_normality(stay)
```

```python
check_normality(leave)
```

комментарий:
 - распределение данных не является нормальным, можно использовать непараметрический тест, такой как тест Манна-Уитни. Тест Манна-Уитни позволяет сравнивать средние значения двух независимых выборок без предположения о нормальности распределения.


#### Проверка гипотез

Гипотезы:

 - Н0 - доход клиентов, которые оттекли и остались, одинаков.
 - H1 - доход клиентов, которые оттекли и остались, различается.

```python
# проведем u-test

_,p_value = st.mannwhitneyu(stay,leave)

alpha = 0.05

print('p-значение:', p_value)

if p_value > alpha:
    print("Нет статистически значимого различия между выборками")
else:
    print("Существует статистически значимое различие между выборками")
```

комментарий:
 - Удалось опровергнуть нулевую гипотезу,что доход клиентов, которые оттекли и остались, одинаков.Рзультаты двух выборок отличаются не случайно,есть основания считать, что эти различия действительно существуют и не объясняются случайными факторами

```python
# проведем t-test

t_statistic, p_value = ttest_ind(stay, leave)

alpha = 0.05

print('p-значение:', p_value)

if p_value > alpha:
    print("Нет статистически значимого различия между выборками")
else:
    print("Существует статистически значимое различие между выборками")
```

комментарий:
 - t-test также неопроверг нулевую гипотезу,что доход клиентов, которые оттекли и остались, одинаков.

В данном случае предпочтительно операться на результаты u-test так как мы выявили, что распределение данных не является нормальным, ведь t-test имеет большую мощность (способность обнаруживать статистические различия) при выполнении предположения о нормальности распределения. t-test оценивает различие между средними значениями двух выборок и оценивает когда данные подчиняются нормальному распределению. Но как сказано было ранне с помощью теста Шапиро-Уилка мы определили, что распределение данных не является нормальным, поэтому u-test более предпочтителен.
